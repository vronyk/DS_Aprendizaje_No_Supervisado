# Modelo predictivo aplicando el algoritmo K-Means utilizando lenguaje Python para resolver un problema de clusterización.

```python
#Loading libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances
from sklearn import preprocessing 
from mpl_toolkits.mplot3d import Axes3D
```

Resumen estadístico de sus datos.

```python
#Read de data
data = pd.read_csv("clientes.csv")
df = pd.DataFrame(data)
print('Dimensiones del df:', df.shape)

# Information on DF types and nulls
df.drop_duplicates(inplace = True)
tab_info=pd.DataFrame(df.dtypes).T.rename(index={0:'tipo de la columna'})
tab_info=tab_info.append(pd.DataFrame(df.isnull().sum()).T.rename(index={0:'campos nulos (cant)'}))
tab_info=tab_info.append(pd.DataFrame(df.isnull().sum()/df.shape[0]*100).T.
                         rename(index={0:'campos nulos (%)'}))
display(tab_info)

display(df[:5])
print(df.describe())
print(df.info())
#Records by category
print(df.groupby('categoria').size())
```

## Graficos

```python
df.drop(['categoria'],1).hist()
plt.show()
```

Relación/grafico de las variables: op, ex y ag, además de las categorías.

```python
#Graph of the variables op, ex y ag
sb.pairplot(df.dropna(), hue='categoria',size=4,vars=["op","ex","ag"],kind='scatter')

#First we define the inputs
X = np.array(df[["op","ex","ag"]])
y = np.array(df['categoria'])
X.shape

#Graph of the variable categorías
fig = plt.figure()
ax = Axes3D(fig)
colores=['blue','red','green','blue','cyan','yellow','orange','black','pink','brown','purple']
asignar=[]
for row in y:
    asignar.append(colores[row])
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
plt.tight_layout()
plt.show()
```

Determinar cuántos k o clúster son adecuados para esta consulta.
(Son necesarios 6 clusters)

```python
distancias = []
k_valores = range(2, 10)
for k in k_valores:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    distancias.append(kmeans.inertia_)

# Graficar la suma de las distancias al cuadrado en función de k
plt.plot(k_valores, distancias, 'bx-')
plt.xlabel('Número de Clústeres (k)')
plt.ylabel('Suma de Distancias al Cuadrado')
plt.title('Método del Codo')
plt.show()
```

Modelo de k means con los cluster definidos en el punto 4.

```python
#We run the algorithm for 6 clusters and get the labels and centroids.

kmeans = KMeans(n_clusters=6, random_state=42).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)
```

Representación gráfica de los cluster.

```python
# Predicting the clusters
labels = kmeans.fit_predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
colores=['red','green','blue','cyan','yellow','brown']
asignar=[]
for row in labels:
    asignar.append(colores[row])
 
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores, s=1000)
```
