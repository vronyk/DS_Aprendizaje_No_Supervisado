4. Considere las siguientes observaciones:

    

    Sin escalar las variables, describa tres iteraciones del algoritmo K-means para k = 2. Use los centroides C1 = (47.5, 37.5, 21.8) y C2 = (53.2, 22.4, 75.3).

```python
import numpy as np
#Definición de puntos y centroides
x = [10,8,34,9,46,68,80,50]
y = [4,99,44,50,77,30,25,35]
z = [50,31,86,57,75,14,40,60]

data = np.array(list(zip(x, y, z)))

c1 = np.array([47.5,37.5,21.8])
c2 = np.array([53.2,22.4,75.3])

centroides = np.array([c1,c2])
```

```python
#Iteración 1
#Cálculo distancia euclidiana

def euclidean_distance(punto, centroide):
    return np.sqrt(np.sum((punto - centroide)**2))

distancias = []
puntos = []
for punto in data:
    distance = [euclidean_distance(punto, centroide) for centroide in centroides]
    punto = punto
    distancias.append(distance)
    puntos.append(punto)
```

```python
# Asignar cada punto a un centroide

clusters = [[] for _ in range(len(centroides))]

for i in range(len(puntos)):
    distancia = distancias[i]
    punto = puntos[i]
    centroide_asignado = np.argmin(distancia)
    clusters[centroide_asignado].append(punto)

for i, cluster in enumerate(clusters):
    print(f"Centroide {i+1}: {cluster}")
```

```python
#Cálculo de nuevos centroides
new_centroids = []

for cluster in clusters:
    new_centroid = np.mean(cluster, axis=0)
    new_centroids.append(new_centroid)
        
# Imprimir nuevos centroides
for i, centroid in enumerate(new_centroids):
    print(f"Nuevo centroide {i+1}: {centroid}")
```

```python
#Iteración 2
#Cálculo distancia euclidiana

distancias = []
puntos = []
for punto in data:
    distance = [euclidean_distance(punto, centroide) for centroide in new_centroids]
    punto = punto
    distancias.append(distance)
    puntos.append(punto)
```

```python
# Asignar cada punto a un centroide

clusters_2 = [[] for _ in range(len(new_centroids))]

for i in range(len(puntos)):
    distancia = distancias[i]
    punto = puntos[i]
    centroide_asignado = np.argmin(distancia)
    clusters_2[centroide_asignado].append(punto)

for i, cluster in enumerate(clusters_2):
    print(f"Centroide {i+1}: {cluster}")
```

```python
#Cálculo de nuevos centroides
new_centroids2 = []

for cluster in clusters_2:
    new_centroid2 = np.mean(cluster, axis=0)
    new_centroids2.append(new_centroid2)
        
# Imprimir nuevos centroides
for i, centroid in enumerate(new_centroids2):
    print(f"Nuevo centroide {i+1}: {centroid}")
```

```python
#Iteración 3
#Cálculo distancia euclidiana

distancias = []
puntos = []
for punto in data:
    distance = [euclidean_distance(punto, centroide) for centroide in new_centroids2]
    punto = punto
    distancias.append(distance)
    puntos.append(punto)
```

```python
# Asignar cada punto a un centroide

clusters_3 = [[] for _ in range(len(new_centroids2))]

for i in range(len(puntos)):
    distancia = distancias[i]
    punto = puntos[i]
    centroide_asignado = np.argmin(distancia)
    clusters_3[centroide_asignado].append(punto)

for i, cluster in enumerate(clusters_3):
    print(f"Centroide {i+1}: {cluster}")
```

```python
#Cálculo de nuevos centroides
new_centroids3 = []

for cluster in clusters_3:
    new_centroid3 = np.mean(cluster, axis=0)
    new_centroids3.append(new_centroid3)
        
# Imprimir nuevos centroides
for i, centroid in enumerate(new_centroids3):
    print(f"Nuevo centroide {i+1}: {centroid}")
```

```python
print("Puntos asignados a cada centroide:")
for i, cluster in enumerate(clusters_3):
    print(f"Cluster {i+1}:")
    for point in cluster:
        print(point)
```

5. Enuncie al menos tres diferencias entre el análisis factorial y el método de componentes principales.

1. El análisis factorial busca identificar variables latentes y PAC busca construir nuevas variables no correlacionadas, representando los datos en un espacio de menor dimensión.
2. El análisis factorial asume una relación lineal entre las variables latentes y las observadas. En el caso del PAC no asume ninguna relación.
3. El análisis factorial requiere que los datos sean cuantitativos u ordinales, en el caso del PAC recibe distintos tipos de datos.
4. En en análisis factorial se puede interpretar los factores extraídos en términos de las variables observadas, no así en el caso del PAC.

6. ¿Qué significa que el método de clusterización sea jerárquico?

En un enfoque jerárquico, el proceso de agrupamiento se lleva a cabo iterativamente, fusionando o dividiendo los clusters existentes.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('evaluacion_mkt_campaign.csv', sep=',')
df
```

```python
# Setear ID como indice
df.set_index('ID', inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
columnas_compras = ['NumDealsPurchases','NumStorePurchases', 'NumWebPurchases', 'NumCatalogPurchases']
df[columnas_compras].describe()
```

```python
# sumar las columans de compras
df['TotalPurchases'] = df[columnas_compras].sum(axis=1)
df['TotalPurchases'].describe()
```

```python
# graficar boxplot de TotalPurchases y histograma
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].boxplot(df['TotalPurchases'])
axs[0].set_title('Boxplot TotalPurchases')
axs[1].hist(df['TotalPurchases'], bins=20)
axs[1].set_title('Histograma TotalPurchases')
plt.show()
```

```python
#filtrar el 5% de datos anomalos de TotalPurchases
q = df['TotalPurchases'].quantile(0.95)
df_filtrado = df[df['TotalPurchases'] < q]
df_filtrado
```

```python
# graficar boxplot de TotalPurchases y histograma
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].boxplot(df_filtrado['TotalPurchases'])
axs[0].set_title('Boxplot TotalPurchases')
axs[1].hist(df_filtrado['TotalPurchases'], bins=20)
axs[1].set_title('Histograma TotalPurchases')
plt.show()
```

```python
# genear X con drop de TotalPurchases, Income
X = df_filtrado.drop(['TotalPurchases', 'Income'], axis=1)
y = df_filtrado['Income']
```

```python
from sklearn.decomposition import PCA
```

```python
# estandarizar los datos
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
pca = PCA(n_components=0.9)
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)

n_components = X_pca.shape[1]
```

```python
print(f'Number of components: {n_components}')
print(f'Explained variance: {pca.explained_variance_ratio_.sum()}')
```

```python
# import k-means
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
k_max = 8  # Número máximo de clusters a probar
wss = np.zeros(k_max)

for k in range(1, k_max + 1):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X_pca)
    wss[k-1] = kmeans.inertia_
wss/=wss.max() # Normalizamos los valores de WSS

plt.figure()
plt.plot(range(1, k_max + 1), wss, color='black')
plt.scatter(range(1, k_max + 1), wss, color='blue')
plt.title("'Elbow Method' - data scaled")
plt.xlabel("Number of clusters K")
plt.ylabel("Total WSS")
plt.show()
```

```python
# range of cluster numbers to try
k_max = 8
min_clusters = 2
max_clusters = 8

# Sum of Squared Errors and silhouette scores
sse = np.zeros(k_max)
silhouette_scores = np.zeros(k_max)

for k in range(min_clusters, max_clusters + 1):

    print(f'Clustering with {k} clusters...')

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10,
    )
    
    kmeans.fit(X_pca)
    
    # Calculate SSE (Sum of Squared Errors)
    sse[k-1] = kmeans.inertia_
    
    # Calculate Silhouette score
    labels = kmeans.labels_
    score = silhouette_score(X_pca, labels)
    silhouette_scores[k-1] = score
```

```python
silhouette_scores.shape , sse.shape , wss.shape
```

```python
plt.figure()
plt.plot(range(1, max_clusters + 1), silhouette_scores, color='black')
plt.scatter(range(1, max_clusters + 1), silhouette_scores, color='blue')
plt.title("'Elbow Method' - data scaled")
plt.xlabel("Number of clusters K")
plt.ylabel("Total Silhouette score")
plt.show()
```

Tanto utilizando SSE y Silhouette Score, se evidencia que el número de clusters a considerar es 3

```python
#Crear instancia de kmeans
k=3
kmeans = KMeans(
    n_clusters=k,
    n_init='auto',
    random_state=42
)

#Entrenar modelo
kmeans.fit(X_pca)

# Obtener centroides
centroides_pca = kmeans.cluster_centers_
```

```python
#Aplicar transformación inversa a pca
centroids_original = pca.inverse_transform(centroides_pca)
print(centroids_original)
```

```python
#Obtener los máximos
cluster = ["Cluster_1", "Cluster_2", "Cluster_3"]
variables = X.columns.to_list()
df_centroides = pd.DataFrame(dict(zip(cluster, centroids_original)), index=variables).abs()

max_index = df_centroides.idxmax()
max_value = df_centroides.max()

max_df = pd.DataFrame({'Variable': max_index, 'Centroide': max_value})

print("Valores Máximos Absolutos Centroides: ")
print()
print(max_df)
```

```python
# Obtener las etiquetas de los clusters asignados a cada punto
labels = kmeans.labels_

# Calcular la desviación estándar de cada variable dentro de cada cluster
std_dev = []
for i in range(k):
    cluster_points = X[labels == i]  # Puntos en el cluster i
    cluster_std = np.std(cluster_points, axis=0)  # Desviación estándar de las variables en el cluster i
    std_dev.append(cluster_std)

# Crear un dataframe con las variables y su desviación estándar dentro de cada cluster
std_dev_df = pd.DataFrame(std_dev, columns=X.columns)
std_dev_df.index.name = 'Cluster'

max_index = std_dev_df.T.idxmax()
max_value = std_dev_df.T.max()

max_df = pd.DataFrame({'Variable': max_index, 'sd': max_value})

print("Variables con mayor desviación estándar para cada cluster: ")
print()
print(max_df)
```

Al considerar los valores absolutos de los centroides, podemos ver que para el Cluster1 la variable más importante es *NumDealsPurchases*, para el Cluster2 *NumStorePurchases*  y para el cluster3 *MntMeatProducts*. Sin embargo, como estos representan distancias, puede que no sea el método más adecuado para ello. Por eso que considerando la desviación estándar de cada cluster, tienen mayor relavancia para el Cluster1 y Cluster3 *MntWines* y para el Cluster2 *Seniority*
